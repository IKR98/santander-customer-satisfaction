# Group some features based name

```python
# features except TARGET
features = [f for f in customers.column_names() if f != 'TARGET']

features
```

## Get the dictionary of the first piece of the features (e.g, imp, ind, num, etc...)

```python
feature_dictionary = {}
for f in features:
    if f.split('_')[0] in feature_dictionary:
        feature_dictionary[f.split('_')[0]] +=1
    else:
        feature_dictionary[f.split('_')[0]] = 1
```
